```python
cat_cols=f.select_dtypes(include="object").columns.tolist()
num_cols=f.select_dtypes(include=["float64","int64"]).columns.tolist()
print(cat_cols)
print(num_cols)
```

```python
for col in cat_cols:
  plt.figure()
  sns.countplot(data=f,x=col,hue="Churn")
  plt.xticks(rotation=45)
  plt.title(f"Churn by {col}")
  plt.tight_layout()
  plt.show()
```

```python
for col in num_cols:
  plt.figure()
  sns.histplot(f[col],kde=True)
  plt.title(f"Distribution of {col}")
  plt.tight_layout()
  plt.show()
```
